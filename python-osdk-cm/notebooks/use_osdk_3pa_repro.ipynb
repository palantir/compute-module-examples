```python
HOSTNAME = "<INSERT STACK URL>"
CLIENT_ID = os.environ["CLIENT_ID"]
CLIENT_SECRET = os.environ["CLIENT_SECRET"]

print(f"CLIENT_ID: {CLIENT_ID} CLIENT_SECRET: {CLIENT_SECRET[:8]}")
```

```python
ConfidentialClientAuth.scopes.extend(["api:ontologies-write","api:ontologies-read"])
```

```python
ConfidentialClientAuth.scopes
```

```python
auth = ConfidentialClientAuth(
    client_id=CLIENT_ID,
    client_secret=CLIENT_SECRET,
    hostname=HOSTNAME,
    should_refresh=True,
)
auth.sign_in_as_service_user()

client = FoundryClient(auth=auth, hostname=HOSTNAME)
```
